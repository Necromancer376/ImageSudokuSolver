```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import easyocr
```

```python
img_path='sudoku3.png'
# img_path='image.jpeg'
```

```python
img_temp = cv2.imread(img_path)
size = 2500
image_resized = cv2.resize(img_temp, (size, size))
cv2.imwrite(img_path, image_resized)
plt.imshow(image_resized)
```

```python
reader=easyocr.Reader(['en'],gpu=False)
res=reader.readtext(img_path)
res
```

```python
img=cv2.imread(img_path)
for i in range(len(res)):
    top_left=tuple(res[i][0][0])
    bottom_right=tuple(res[i][0][2])
    img=cv2.rectangle(img,top_left,bottom_right,[0,0,255],3)
plt.imshow(img)
img.shape
```

```python
#RETR_TREE
grid=cv2.imread(img_path)
grid_copy=grid.copy()
grid_gray=cv2.cvtColor(grid,cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(grid_gray, 190, 255,cv2.THRESH_BINARY_INV)
contours,heirarchy= cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
image=cv2.drawContours(grid_copy,contours,-1,(0,255,0),2)
plt.figure(figsize=[7,7])
plt.subplot(121)
plt.imshow(image)
plt.axis('off')
plt.subplot(122)
plt.imshow(grid)
plt.axis('off')
```
